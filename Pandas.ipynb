# Usando pandas para escrever e ler arquivos Excel

Você precisará de alguns os seguintes pacotes:
   
    xlwt: para escrever arquivos .xls
    openpyxl ou XlsxWriter: para escrever arquivos .xlsx
    xrld: para ler arquivos Excel
    
Você pode instalar via pip:
    
    pip install xlwt openpyxl xlsxwriter xlrd
    
ou pelo conda:
    
    conda install xlwt openpyxl xlsxwriter xlrd
    
Obs: Você não precisa instalar todos esses pacotes, caso só use o xls instale apenas os pacotes para o xls

## Lendo e escrevendo arquivos Excel:

Você pode escrever arquivos Excel com o comando `.to_excel( )`:

    .to_excel('nome_do_arquivo.xlsx') # Ou o caminho para o diretório escolhido
    
Você pode ler arquivos Excel com o comando `.read_excel( )`:

    .read_excel('nome_do_arquivo.xlsx') # Ou o caminho para o diretório escolhido

```python
df.to_excel('data.xlsx')
```

```python
df = pd.read_excel('data.xlsx', index_col=0)
df
```
